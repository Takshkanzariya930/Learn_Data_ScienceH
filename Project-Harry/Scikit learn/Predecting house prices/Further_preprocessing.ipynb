```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
strat_train_set = pd.read_csv("strat_train_set.csv")
strat_test_set = pd.read_csv("strat_test_set.csv")
```

```python
housing = strat_train_set.drop("median_house_value", axis=1)
housing_lables = strat_train_set["median_house_value"].copy()
```

```python
housing
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy = "median")
```

```python
housing_num = housing.select_dtypes(include=[np.number])
```

```python
housing_num
```

```python
imputer.fit(housing_num)
```

```python
housing_imputed = imputer.fit_transform(housing_num)
```

```python
housing_non = pd.DataFrame(housing_imputed,columns=housing_num.columns,index=housing_num.index)
```

```python
housing_non.info()
```

```python
housing_non["ocean_proximity"] = housing["ocean_proximity"]
```

```python
housing_non
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
onehot_encoder = OneHotEncoder()
```

```python
housing_cat = housing_non[['ocean_proximity']]
housing_cat
```

```python
set(housing_cat['ocean_proximity'])
```

```python
housing_cat = onehot_encoder.fit_transform(housing_cat)
```

```python
housing_cat.toarray()
```

```python
housing_cat = pd.DataFrame(housing_cat.toarray(),columns=['<1H OCEAN', 'INLAND', 'ISLAND', 'NEAR BAY', 'NEAR OCEAN'],index=housing_non.index)
```

```python
housing_cat
```

```python
housing_non.drop("ocean_proximity",axis=1,inplace=True)
```

```python
housing_non = pd.concat([housing_non, housing_cat], axis=1)
```

```python
housing_non
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler = MinMaxScaler(feature_range = (-1,1))
```

```python
housing_non_minmax_scaled = min_max_scaler.fit_transform(housing_non)
```

```python
housing_non_minmax_scaled = pd.DataFrame(housing_non_minmax_scaled,columns=housing_non.columns,index=housing_non.index)
```

```python
housing_non_minmax_scaled
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scaler = StandardScaler()
```

```python
housing_std_scaled = std_scaler.fit_transform(housing_non)
```
