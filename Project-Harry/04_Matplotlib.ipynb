```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Years = [ 2011 + x for x in range(10)]
```

```python
kohli = [ 752,  923, 1181, 1585, 1192, 1740, 1947, 1447, 1538,  787]
rohit = [1396, 1683, 1891, 1671, 1888, 1330,  668,  966, 1979, 1209]
sehwag = [1000, 1887, 1769,  631, 1455, 1597, 1671,  727, 1652, 1754]
```

```python
plt.plot(Years,kohli,label = "Kohli",color = "green")
plt.plot(Years,rohit,label = "Rohit",color = "red",linestyle = "--")
plt.plot(Years,sehwag,label = "Sehwag",color = "blue",linestyle = "-.")
plt.legend()

plt.xlabel("Years")
plt.ylabel("Run scored")
plt.title("Performance")
plt.grid("true")
plt.style.use('bmh')
plt.tight_layout()
```
