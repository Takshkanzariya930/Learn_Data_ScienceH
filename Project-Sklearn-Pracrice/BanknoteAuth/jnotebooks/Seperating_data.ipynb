```python
import pandas as pd
```

```python
data = pd.read_csv("../data_banknote_authentication.csv")
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
```

```python
for train_index, test_index in split.split(data,data["Y"]):
    strat_train_set = data.loc[train_index]
    strat_test_set = data.loc[test_index]
    
```

```python
import matplotlib.pyplot as plt
```

```python
data["Y"].value_counts().sort_index().plot.bar(rot=0, grid=True)  
plt.title("Authenticity Destribution") 
plt.xlabel("Authentic")
plt.ylabel("Number of Instances")
```

```python
strat_train_set["Y"].value_counts().sort_index().plot.bar(rot=0, grid=True)  
plt.title("Authenticity Destribution") 
plt.xlabel("Authentic")
plt.ylabel("Number of Instances")
```

```python
strat_test_set["Y"].value_counts().sort_index().plot.bar(rot=0, grid=True)  
plt.title("Authenticity Destribution") 
plt.xlabel("Authentic")
plt.ylabel("Number of Instances")
```

```python
strat_test_set.to_csv("strat_train_set.csv", index=False)
```
